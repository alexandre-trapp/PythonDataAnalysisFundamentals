```python
#criando o banco de dados twitterdb
db = client.twitterdb
```

```python
# criando a collection "col"
col = db.tweets
```

```python
# criando dataset com dados retornados do mongodb
dataset = [ {"created_at": item["created_at"], "text": item["text"]} for item in col.find()]
```

```python
# importando modulo pandas p/ trabalhar com datasets em python
import pandas as pd
```

```python
# criando um dataframe a partir do dataset
df = pd.DataFrame(dataset)
```

```python
#imprimindo o dataframe
df
```

```python
# analise de dados com pandas e Scikit-Learn
#importando o modulo Scikit Learn

from sklearn.feature_extraction.text import CountVectorizer
```

```python
#usando o metodo CountVectorizer para criar uma matriz de documentos
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
# contando o numero de ocorrencias das principais palavras em nosso dataset do twitter
word_count = pd.DataFrame(cv.get_feature_names(), columns=["word"])
word_count["count"] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values("count", ascending=False).reset_index(drop=True)
word_count[:100]
```
